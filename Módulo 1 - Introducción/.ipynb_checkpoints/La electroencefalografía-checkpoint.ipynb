### Introducción

El electroencefalograma (EEG) es una indispensable y útil herramienta con mas de 85 anos de desarrollo que realiza un registro de las oscilaciones del potencial eléctrico del cerebro adquirido mediante electrodos colocados sobre el cuero cabelludo (Nunez & Srnivasan, 2006)[1]. En la actualidad, la electroencefalografía es una técnica altamente utilizada en los estudios neurológicos y psiquiátricos (Boutros, 2013)[2],debido a que permite identificar la actividad cerebral y el comportamiento mediante la detección de senales eléctricas, además se ha utilizado para la investigación de las corrientes y oscilaciones intrínsecas de voltaje extra celulares (Buzsaki, et. al. 2003)[3]. Con la llegada de la era digital y su integración con el EEG, esta técnica se ha sofisticado para hacer uso de las distintas ciencias de la computación y análisis de datos extendiendo su uso y aplicación más allá del laboratorio permitiendo realizar una investigación robusta sobre las distintas ondas cerebrales, neuropatologáa y diagnóstico clínico (Feyissa & Tatum, 2019)[4].

### Obtención de los datos

La técnica se enfoca en la medición, amplificación y registro de las diferencias de potencial entre los campos oscilantes en el tiempo del cerebro, estos registros se estudian dentro de el ancho de banda de 0.16-100 [Hz]. La medición de la actividad eléctrica consta de 4 fases principales: adquisición, amplificación, filtrado y procesamiento, y digitalización. 
Una muestra o registro de EEG está caracterizada por su duración, ubicación del registro y frecuencia de muestreo a la que fue obtenida.

Hasta el momento se cuenta con la visualización temporal de la señal completa pero usualmente es útil poder desplegar segmentos de la señal específicos por lo que definiremos una función que nos extraiga dichos segmentos como nuevas series de tiempo.

```python
def s_cut(s,t_i,t_f,fm=500):
    '''
    s:= señal a recortar
    t_i:= tiempo inicial en segundos 
    t_f:= tiempo final en segundos
    fm:= frecuencia de muestreo'''
    N = len(s)
    s_time = N/fm #Duración de la señal
    time = np.linspace(0,s_time,N) #Vector de tiempo completo 
    time_frag = time[int(t_i*fm):int(t_f*fm)] #Vector de tiempo recortado
    s_frag = s[int(t_i*fm):int(t_f*fm)] #señal en el tiempo deseado
    return time_frag,s_frag #La función regresa ambos vectores en el intervalo de tiempo real deseado
```

```python
fig,ax = plt.subplots(3,1,figsize=(15,10))

#Definimos un tiempo inicial y un final para cada gráfica y aplicamos la función
ti  = 0.5 
tf = 1.5
t1,s1 = s_cut(EEG,ti,tf)

ax[0].plot(t1,s1*1000)
ax[0].set_ylabel('Amplitud $mV$')
ax[0].set_ylim((-100,100))
ax[0].set_title(f'$S_t @ [{ti},{tf}]$')
ax[0].grid()

ti  = 2
tf = 2.7
t1,s1 = s_cut(EEG,ti,tf)
ax[1].plot(t1,s1*1000)
ax[1].set_ylabel('Amplitud $mV$')
ax[1].set_ylim((-100,100))
ax[1].set_title(f'$S_t @ [{ti},{tf}]$')
ax[1].grid()

ti  = 0.52
tf = 0.679
t1,s1 = s_cut(EEG,ti,tf)
ax[2].plot(t1,s1*1000)
ax[2].set_xlabel('Tiempo [s]')
ax[2].set_ylabel('Amplitud $mV$')
ax[2].set_ylim((-100,100))
ax[2].set_title(f'$S_t @ [{ti},{tf}]$')
ax[2].grid()
```

#### Visualización Frecuencial

Para el análisis e inspección de series de tiempo de bioseñales, siempre es útil complementar la visualización temporal con la frecuencial, para lograrlo es necesario implementar la transformada de Fourier; es decir expresamos la señal en el dominio de la frecuencia. <br>
Matemáticamente la transformada de Fourier de una función contínua se expresa de la siguiente manera:
<br>
<br>
<div align="center">$$\mathcal{F}\{g(t)\} = \hat{g}(t)= \frac{1}{\sqrt{2\pi}}\int^{\infty}_{-\infty} e^{-i\omega t}g(t) dt$$ <div>
<br> 
y su transformada inversa:
<br>
<br>
$$\mathcal{F}^{-1}\{ \hat{g}(t) \} = g(t) = \frac{1}{\sqrt{2\pi}}\int^{\infty}_{-\infty} e^{i\omega t}\hat{g}(t) d\omega$$

En este caso al tratarse de señales discretas se emplearía la definición discreta de la transformada: <br> <br>
<div align="center">$\mathbf{X_k} = \sum^{N-1}_{n=0} x_n e^{-\frac{2\pi i}{N}kn}$ Tal que el ínidce $k=0,\cdots,N-1$ <br><br>
$x_n = \frac{1}{N} \sum^{N-1}_{n=0} \mathbf{X_k}e^{\frac{2\pi i}{N}kn}$ la transformada inversa
</div>
<br>
Para calcular la transformada se hace uso del algoritmo más eficiente, la transformada rápida de Fourier FFT.

```python
x = EEG.to_numpy() *1000 #Renombramos la variable
N = len(x)
signal_fft = sc.fft(x-x.mean()) #Aplicamos la FFT de la paq. de Scipy
T = 1.0/500 #Calculamos el periodo de la señal como el recíproco de la frecuencia de muestreo
frequen = sc.fft.fftfreq(N,T)[1:N//2] #Construimos el vector de frecuencias para graficar la transformada
signal_fft_abs = 2.0/N * np.abs(signal_fft[1:N//2]) #Normalizamos la transformada y eliminanos los valores resultantes de la paridad (freq. neg)
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,4))

ax.plot(frequen,signal_fft_abs,c='green')
ax.set_xlabel('Frecuencia [Hz]')
ax.set_ylabel('Potencia $[mV^2/Hz]$')
ax.set_title('Transformada de Fourier')
ax.set_xlim((0,30))
ax.grid()
```

En esta visualización podemos notar en que anchos de banda se esta concentrando la mayor cantidad de energía de la señal, y a primer vistazo, notamos como coincide con lo esperado observar un pico de energía en el ancho de banda asociado a las ondas alfa de EEG [8,12] Hz. Este resultado es consecuencia de la condición bajo la que se obtuvo el registro y la ubicación espacial del mismo.

Una técnica para enfatizar los ritmos de baja amplitud-potencia en el espectro eclipsados por los ritmos en alta amplitud-potencia es modificar la escala a decibeles, la cual es una escala logarítmica y fácil de calcular:

```python
fig, ax = plt.subplots(1,1,figsize=(15,4))

ax.plot(frequen,20* np.log10(signal_fft_abs/signal_fft_abs.max()),c='red')
ax.set_xlabel('Frecuencia [Hz]')
ax.set_ylabel('Potencia $[dB]$')
ax.set_title('TF en escala decibel')
ax.set_xlim((0,50))
ax.set_ylim((-60,10))
ax.grid()
```

Para cambiar a la escala decibel el procedimiento fue, normalizar el espectro dividiéndolo entre el valor máximo observado y luego calcular el logaritmo base 10 de esta normalización, finalmente se multiplica el resultado por 20. Aquí notamos como el ritmo dominante sigue siendo el de las ondas alfa en el ancho de banda [8-12] Hz.
